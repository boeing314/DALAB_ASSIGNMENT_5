# A5: Visualizing Data Veracity Challenges in Multi-Label Classification

## Part A: Preprocessing and Initial Setup

### Data loading

We are reading the data from yeast.csv file and then seperating the feature variables from the target variables.

```python
import pandas as pd
import numpy as np
#Reading the dataset and and splittting the feature variables and target variables
df=pd.read_csv('yeast.csv')
X=df[[col for col in df.columns if col.startswith('Att')]]
Y=df[[col for col in df.columns if col.startswith('Class')]]
Y = Y.map(lambda x: 1 if x == "b'1'" else 0 if x == "b'0'" else x)
```

### Dimensionality Check:

```python
print("Number of datapoint are        : ",df.shape[0])
print("Number of feature variables are: ",X.shape[1])
print("Number of target variables are : ",Y.shape[1])
```

### Label selection for visualization

We cannot choose all the labels for as 14 colors can be overwhelming for visualization. So we will choose few of the most frequent single-label classes and multi-label classes and we combine the rest of the labels into one category.

```python
class_cols = [f'Class{i}' for i in range(1, 15)]
Y['label_count'] = Y[class_cols].sum(axis=1)
single_label_rows = Y[Y['label_count'] == 1].copy()
multi_label_rows = Y[Y['label_count'] > 1].copy()

single_label_rows['active_class'] = single_label_rows[class_cols].idxmax(axis=1)
top2_single = (single_label_rows['active_class'].value_counts().nlargest(2).index.tolist())

multi_label_rows['combo'] = (multi_label_rows[class_cols].apply(lambda row: ' + '.join(sorted([col for col in class_cols if row[col] == 1])), axis=1))
top1_multi = multi_label_rows['combo'].value_counts().nlargest(1).index.tolist()

Y.drop(columns='label_count', inplace=True)

print("Top 2 single-label classes:", top2_single)
print("Most frequent multi-label combo:", top1_multi)

```

We see that 'Class 1' is the only single-label class while 'Class12 + Class13 + Class3 + Class4' is the most frequent multilabel class.

So based on this we will create a new target variable for coloring that represents the most frequent single-label classes and the most frequent multi-label combination. We will assign an 'others' category to the rest.

```python
Y_simplified = pd.DataFrame(np.zeros((2417, 3), dtype=int), columns=['Class 1', 'Class12 + Class13 + Class3 + Class4', 'others'])

single_label_mask = (Y['Class1'] == 1) & (Y.drop(columns=['Class1']).sum(axis=1) == 0)
Y_simplified.loc[single_label_mask, 'Class 1'] = 1

multi_label_mask1 = (Y[['Class3', 'Class4', 'Class12', 'Class13']].sum(axis=1) == 4) & (Y.sum(axis=1)==4)
Y_simplified.loc[multi_label_mask1, 'Class12 + Class13 + Class3 + Class4'] = 1

others_mask = ~(single_label_mask | multi_label_mask1 )
Y_simplified.loc[others_mask, 'others'] = 1

print(Y_simplified.sum())
```

### Scaling

Scaling is important before applying distance-based dimensionality reduction techniques like t-SNE and Isomap because these methods rely heavily on distance metrics to reduce the dimensionality of the data. If the features in the dataset have different scales, the features with larger numerical ranges will dominate the distance calculation. This biasness will distort the results of the dimensionality reduction.

So we standardize the data so that each input feature has an average of 0 and variance of 1.

```python
from sklearn.preprocessing import StandardScaler
# Standarizing
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

## Part B: t-SNE and Veracity Inspection

### t-SNE Implementation

t-Distributed Stochastic Neighbor Embedding (t-SNE) is a non-linear dimensionality reduction technique that projects high-dimensional data into a lower dimension while preserving local neighborhood relationships. It converts pairwise distances into probabilities and minimizes the Kullbackâ€“Leibler divergence between high- and low-dimensional similarity distributions.

We will vary the perplexity parameter to check which gives the best result. The perplexity parameter controls the balance between local and global structure. The lower values of perplexity parameter captures local structure while higher values capture global structures.

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

#Different perplexity values to plot for
perplexities = [5,10,15,20,25,30,35,40,45,50,55,60]

class_names = ['Class 1', 'Class3+Class4+Class12+Class13', 'Others']
y_labels = np.argmax(Y_simplified, axis=1)
class_colors = ["#300de0", "#950909", "#7fc584"]
#Creatinng subplots
fig, axes = plt.subplots(4, 3, figsize=(15, 14))
axes = axes.flatten()
#plotting the points
for i, perp in enumerate(perplexities):
    tsne = TSNE(n_components=2, perplexity=perp, random_state=0)
    X_tsne = tsne.fit_transform(X_scaled)

    ax = axes[i]
    
    for j, cls in enumerate(np.unique(y_labels)[::-1]):
        idx = y_labels == cls
        ax.scatter(X_tsne[idx, 0], X_tsne[idx, 1],color=class_colors[::-1][j],label=class_names[::-1][j],s=20, alpha=0.8, edgecolor='k', linewidth=0.3)

    ax.set_title(f"Perplexity = {perp}", fontsize=12)
    ax.set_xlabel("t-SNE 1")
    ax.set_ylabel("t-SNE 2")
    ax.grid(True, alpha=0.2)

plt.tight_layout(rect=[0, 0, 1, 0.97])
#legend
handles, labels = ax.get_legend_handles_labels()
fig.legend(handles, labels, title="Classes",loc='lower center', bbox_to_anchor=(0.5, -0.05),ncol=len(class_names), frameon=True)
fig.suptitle("t-SNE Visualization for Various Perplexities", fontsize=20,y=1.05, fontweight='bold')
plt.show()
```

From observing the above plots, we conclude perplexity=45 is the best as it captures a good cluster of top single-label class and the top multi-label class.

### Visualization

We will visualize the case for which perplexity is 45

```python
from matplotlib.lines import Line2D 
plt.figure(figsize=(7,6))

class_names = ['Class 1', 'Class3+Class4+Class12+Class13', 'Others']
y_labels = np.argmax(Y_simplified, axis=1)
class_colors = ["#300de0", "#950909", "#7fc584"]

# Initialize t-SNE with desired parameters
tsne = TSNE(n_components=2, perplexity=45, random_state=0)

# Applying t-SNE transformation
X_tsne = tsne.fit_transform(X_scaled)

# Plot each class separately
for j, cls in enumerate(np.unique(y_labels)[::-1]):
    idx = y_labels == cls
    plt.scatter(X_tsne[idx, 0], X_tsne[idx, 1],color=class_colors[::-1][j],label=class_names[::-1][j],s=40, alpha=0.8, edgecolor='k', linewidth=0.5)

# Set title and labels
plt.title(f"t-SNE Visualization (Perplexity = 45)", fontsize=12)
plt.xlabel("t-SNE 1")
plt.ylabel("t-SNE 2")
plt.grid(True, alpha=0.2)
# Definition to plot an elipse to mark the noisy,outlier and other similar points


legend_elements = [Line2D([0], [0], marker='o', color='w',markeredgecolor=class_colors[i], markerfacecolor=class_colors[i],markeredgewidth='3', markersize=10, label=class_names[i]) 
                   for i in range(len(class_names)-1,-1,-1)]
plt.legend(handles=legend_elements, title="Classes", loc='upper left', bbox_to_anchor=(1.05, 1), frameon=True)
plt.show()
```

### Veracity Inspection

**Noisy/Ambiguous Labels**  
 - These refer points where one color is deeply embedded within a cluster of a different color.
 - Incorrect labelling or experimental error is the most likely reason this happens.
 - It is less likely that this point lies in the boundary of two classes because majority of the adjacent points belong to a different class.

**Outliers**
 - These are isolated points or small clusters far from the main clusters.
 - The small clusters might be rare, unique sample or some extreme sample for that particular category.
 - The isolated points could be due to it being rare or unique sample or it being an experimental error.

**Hard-to-Learn Samples**
 - These are points where the category colors are thoroughly mixed.
 - t-SNE prioritizes local structure so as a result points that are well-separated in high-dimensional space might appear very close in the plot as it does not capture global structure very well.
 - This may also be due to insufficient number of features leading to feature overlap in this case. 

***All these points are marked in the following plot***

```python
from matplotlib.lines import Line2D 

# Create a figure and axis for plotting
plt.figure(figsize=(7,6))

class_names = ['Hard-to-Learn','Outliers','Noisy/Ambiguous','Class 1', 'Class3+Class4+Class12+Class13', 'Others']
y_labels = np.argmax(Y_simplified, axis=1)
class_colors = ["#ecf00bd6","#c71bd06d","#ef7c09dc","#300de0", "#950909", "#7fc584"]
face_colors=["none","none","none","#300de0", "#950909", "#7fc584"]

# Initialize t-SNE with desired parameters
tsne = TSNE(n_components=2, perplexity=45, random_state=0)

# Applying t-SNE transformation
X_tsne = tsne.fit_transform(X_scaled)

# Plot each class separately
for j, cls in enumerate(np.unique(y_labels)[::-1]):
    idx = y_labels == cls
    plt.scatter(X_tsne[idx, 0], X_tsne[idx, 1],color=class_colors[::-1][j],label=class_names[::-1][j],s=40, alpha=0.8, edgecolor='k', linewidth=0.5)

# Set title and labels
plt.title(f"t-SNE Visualization with Veracity Inspection", fontsize=12)
plt.xlabel("t-SNE 1")
plt.ylabel("t-SNE 2")
plt.grid(True, alpha=0.2)
# Definition to plot an elipse to mark the noisy,outlier and other similar points
def plot_ellipse(ax, center, a, b, angle=0, color='black', linewidth=2):
    theta = np.linspace(0, 2 * np.pi, 1000)
    x = a * np.cos(theta)
    y = b * np.sin(theta)
    x_rot = x * np.cos(np.radians(angle)) - y * np.sin(np.radians(angle))
    y_rot = x * np.sin(np.radians(angle)) + y * np.cos(np.radians(angle))
    x_rot += center[0]
    y_rot += center[1]
    ax.plot(x_rot, y_rot, color=color, linewidth=linewidth,alpha=1)

# Marking the noisy points
plot_ellipse(plt.gca(), (21,-27), 2, 3, 0, color="#ef7c09dc", linewidth=2)
plot_ellipse(plt.gca(), (7.5,-3), 2.5, 2.5, 0, color="#ef7c09dc", linewidth=2)
plot_ellipse(plt.gca(), (-36,-4), 2.5, 2.5, 0, color="#ef7c09dc", linewidth=2)
# Marking the outliers
plot_ellipse(plt.gca(), (-11,-35), 2, 2, 0, color="#c71bd06d", linewidth=2)
plot_ellipse(plt.gca(), (31.5,13.5), 3.5, 2, 20, color="#c71bd06d", linewidth=2)
plot_ellipse(plt.gca(), (36.5,3.5), 4, 2, 45, color="#c71bd06d", linewidth=2)
# Marking the hard to learn points
plot_ellipse(plt.gca(), (-27,-7), 4, 4, 0, color="#ecf00bd6", linewidth=2)
plot_ellipse(plt.gca(), (7.5,5), 6, 5, 0, color="#ecf00bd6", linewidth=2)
plot_ellipse(plt.gca(), (-30,15.5), 6, 5, -30, color="#ecf00bd6", linewidth=2)
plt.tight_layout(rect=[0, 0, 1, 0.97])

custom_markers=['s','s','s','o','o','o']
edge_W=['3','3','3','3','3','3']
legend_elements = [Line2D([0], [0], marker=custom_markers[i], color='w',markeredgecolor=class_colors[i], markerfacecolor=face_colors[i],markeredgewidth=edge_W[i], markersize=10, label=class_names[i]) 
                   for i in range(len(class_names)-1,-1,-1)]
plt.legend(handles=legend_elements, title="Classes", loc='upper left', bbox_to_anchor=(1.05, 1), frameon=True)
plt.show()
```

## Part C: Isomap and Manifold Learning

### Isomap Implementation

Isomap preserves global structure while t-SNE preserves local structure.

**Isomap**
 - Isomap focuses on maintaining the global geometry of the data manifold by computing the geodesic distances (shortest paths along the data manifold) between all pairs of points and it tries to preserve these distances in the lower dimensional space.
 - Because of this Isomap maintains the overall shape, continuity and large-scale relationships between distant clusters.

**t-SNE**
 - It models pairwise similarities as probabilities and minimizes the difference between high and low dimensional neighbor distributions.
 - It emphasizes local neighborhoods by keeping nearby points close but it intentionally distorts global distances to reveal tight clusters.

```python
from sklearn.manifold import Isomap
#Isomap implementation
isomap = Isomap(n_components=2, n_neighbors=10)
X_iso = isomap.fit_transform(X_scaled)
```

**Observation**
- In t-SNE we can clusters are tightly packed and seperable when compared to Isomap for which points are slightly spread out. 
- This suggests t-SNE captures local structure while Isomap captures global structure.

**Data Manifold**
- A data manifold is a lower-dimensional, continuous surface embedded within a high-dimensional space where the data actually lies.
- Isomap and t-SNE aim to uncover and flatten this underlying structure for visualization or learning.

**Observtion from the Isomap plot**
- The points are spread out but clustered suggesting that the underlying manifold is nonlinear and somewhat curved.

- There appear to be distinct clusters but with overlaps which means that different classes share nearby regions on the manifold.

- Isomap tries to preserve global geometry. The fact that clusters are spread yet overlapping suggests a slightly complex manifold not extremely tangled but not flat either.

**The more curved or complex the manifold the harder it is for classifiers to separate classes**. In this case it is a slightly complex manifold so it is difficult for the classifier to classify the points as this complexity can lead to overlapping decision boundaries.
